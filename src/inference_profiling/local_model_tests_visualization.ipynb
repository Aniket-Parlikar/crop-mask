```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('local_model_tests.csv')
```

```python
df.tail()
```

```python
df['Batch_size']=df['Batch_size'].astype('str')
```

```python
Namibia_df=df.query("Model_name=='Namibia_North_2020'")
Malawi_df=df.query("Model_name=='Malawi_2020_September'")
```

```python
#Plotting the boxplot of eeg sensor related columns
figure,axs = plt.subplots(1,2, figsize=(10, 3), facecolor='w', edgecolor='k')
figure.suptitle('Peak cpu and memory percentage variation according to the batch size')

axs = axs.flatten()
axs[0].plot(Namibia_df['Batch_size'],Namibia_df['Peak_cpu_percent'])
axs[1].plot(Namibia_df['Batch_size'],Namibia_df['Peak_memory_percent'])
```

```python
#Plotting the boxplot of eeg sensor related columns
figure,axs = plt.subplots(1,2, figsize=(10, 3), facecolor='w', edgecolor='k')
figure.suptitle('Peak cpu percentage variation according to the batch size')

axs = axs.flatten()
axs[0].plot(Namibia_df['Batch_size'],Namibia_df['Peak_cpu_percent'])
axs[0].legend(['Namibia_North_2020'])
#axs[0].set_color('yellow')
axs[1].plot(Malawi_df['Batch_size'],Malawi_df['Peak_cpu_percent'],'y')
axs[1].legend(['Malawi_2020_September'])
```

```python
#Plotting the boxplot of eeg sensor related columns
figure,axs = plt.subplots(1,2, figsize=(10, 3), facecolor='w', edgecolor='k')
figure.suptitle('Peak memory percentage variation according to the batch size')

axs = axs.flatten()
axs[0].plot(Namibia_df['Batch_size'],Namibia_df['Peak_memory_percent'])
axs[0].legend(['Namibia_North_2020'])
#axs[0].set_color('yellow')
axs[1].plot(Malawi_df['Batch_size'],Malawi_df['Peak_memory_percent'],'y')
axs[1].legend(['Malawi_2020_September'])
```
